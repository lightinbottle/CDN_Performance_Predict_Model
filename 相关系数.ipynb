```python
import numpy as np
import pandas as pd
import scipy.spatial
import math
from scipy.spatial.distance import correlation
```

```python
a=pd.DataFrame()
num=100
```

```python
a[1]=list(range(num))
```

```python
a[2]=[i**2 for i in range(num)]
```

```python
a[3]=[i**3 for i in range(num)]
a[4]=[math.exp(i+1)+math.log(i+1)+math.sin(i)-i+math.cos(i)+math.sqrt(i) for i in range(num)]
a
```

```python
a.corr()
```

```python
a.corr(method='spearman')
```

```python
correlation(a[1].values ,a[4].values)
```
