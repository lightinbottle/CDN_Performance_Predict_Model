```python
import os
import hashlib
import pandas as pd
```

```python
dir_="./wangsu_result/pptv_20180119_img"
dir_=""
```

```python
result_list=[]
dir_list=os.listdir("./wangsu_result")
for dir_ in dir_list:
    result_list+=os.listdir("./wangsu_result/"+dir_)
```

```python
result_list=list(set(result_list))
for i in range(len(result_list)):
    result_list[i]=result_list[i][0:-4]
```

```python
def MD5(src):
    m1=hashlib.md5()
    m1.update(src.encode("UTF-8"))
    return m1.hexdigest()
```

```python
md5_dict={}
for i in result_list:
    temp=MD5(i)
    md5_dict[temp]=i
```

```python
hit_rate_df2=hit_rate_df.groupby('time',as_index=False).sum()
```
