```python
def get_related(machine, string):
    feature_list = list(machine.columns)
    return [i for i in feature_list if i[0:len(string)] == string]
```

```python
def sum_columns(data, col_list):
    return data[col_list].apply(np.sum, axis=1)
```

```python
related_feature_strings=["cpu","channeltraffic.in","channeltraffic.out","load.total_load",
                          "resptime.resp.resp","iowait.wait","ioutil.util","mem"]
```

```python
def feature_filter(machine):
    feature_list = list(machine.columns)
    related_feature_list=[]
    for string in related_feature_strings:
        related_feature_list=related_feature_list+get_related(machine,string)
    drop_feature_list=list(set(feature_list)-set(related_feature_list))
    machine_droped=machine.drop(drop_feature_list,axis=1)
    return machine_droped
```

```python
machine0=feature_filter(machine0)
machine1=feature_filter(machine1)
machine2=feature_filter(machine2)
machine3=feature_filter(machine3)
```

### 聚合CPU

 当前时刻所有核的使用率按值的大小排序，取值较大的qos_cputop(默认25%)数量的值做平均，再做归一化处理，最后再平滑。例如一台机器有24个核，每个时刻有24个cpu值，则取最大的6个值（24*25%=6，不足整数则补全整数）做平均，再做归一化和平滑处理。

```python
def aggregate_cpu(machine):
    cpu_related = get_related(machine, 'cpu')
    num=len(cpu_related)
    cpu_value=machine[cpu_related].values
    cpu_value.sort(axis=1)
    cpu_value=cpu_value[:,-(int(len(cpu_related)*0.25)):]
    cpu_value=np.sum(cpu_value,axis=1)/int(len(cpu_related)*0.25)
    machine_aggregated = machine.drop(cpu_related, axis=1)
    machine_aggregated["cpu"]=cpu_value
    return machine_aggregated
```

```python
def aggregate_ioutil(machine):
    ioutil_related = get_related(machine, 'ioutil.util')
    num=len(ioutil_related)
    ioutil = sum_columns(machine, ioutil_related)
    machine_aggregated = machine.drop(ioutil_related, axis=1)
    machine_aggregated['ioutil.util'] = ioutil/num
    return machine_aggregated
```

### 聚合resptime.resp.xx

80端口是核心是直接对外提供服务的端口，和达标率直接相关 直接采用  
resptime.resp.8XXX 除了80端口外其他端口的的响应时间均值、极大值

```python
def aggregate_resptime(machine):
    resptime_related = get_related(machine, 'resptime.resp.resp')
    resptime_related.remove('resptime.resp.resp80')
    num=len(resptime_related)-1
    resptime_resp = sum_columns(machine, resptime_related)
    machine_aggregated = machine.drop(resptime_related, axis=1)
    machine_aggregated['resptime.resp.8X.mean'] = resptime_resp/num  #均值
    machine_aggregated['resptime.resp.8X.max']=np.max(machine0[resptime_related].values,axis=1)
    return machine_aggregated
```

```python
def aggregate_mem(machine):
    mem_related = get_related(machine, 'mem')
    print(mem_related)
    machine_aggregated = machine.drop(mem_related, axis=1)
    machine_aggregated["mem.usage"]=(machine["mem.total"]-machine["mem.free"]-machine["mem.bufers"]-machine["mem.cached"])/machine["mem.total"]
    return machine_aggregated
```

### 单机聚合总结

```python
def aggregate_cpu(machine):
    cpu_related = get_related(machine, 'cpu')
    num=len(cpu_related)
    cpu_value=machine[cpu_related].values
    cpu_value.sort(axis=1)
    cpu_value=cpu_value[:,-(int(len(cpu_related)*0.25)):]
    cpu_value=np.sum(cpu_value,axis=1)/int(len(cpu_related)*0.25)
    machine_aggregated = machine.drop(cpu_related, axis=1)
    machine_aggregated["cpu"]=cpu_value
    return machine_aggregated
```

```python
def aggregate_ioutil(machine):
    ioutil_related = get_related(machine, 'ioutil.util')
    num=len(ioutil_related)
    ioutil = sum_columns(machine, ioutil_related)
    machine_aggregated = machine.drop(ioutil_related, axis=1)
    machine_aggregated['ioutil.util'] = ioutil/num
    return machine_aggregated
```

```python
def aggregate_resptime(machine):
    resptime_related = get_related(machine, 'resptime.resp.resp')
    resptime_related.remove('resptime.resp.resp80')
    num=len(resptime_related)-1
    resptime_resp = sum_columns(machine, resptime_related)
    machine_aggregated = machine.drop(resptime_related, axis=1)
    machine_aggregated['resptime.resp.8X.mean'] = resptime_resp/num  #均值
    machine_aggregated['resptime.resp.8X.max']=np.max(machine[resptime_related].values,axis=1)
    return machine_aggregated
```

```python
def aggregate_mem(machine):
    mem_related = get_related(machine, 'mem')
    machine_aggregated = machine.drop(mem_related, axis=1)
    machine_aggregated["mem.usage"]=(machine["mem.total"]-machine["mem.free"]-machine["mem.bufers"]-machine["mem.cached"])/machine["mem.total"]
    return machine_aggregated
```

```python
def aggregate_single_machine(machine):
    aggregated_machine = feature_filter(machine)
    aggregated_machine = aggregate_cpu(aggregated_machine)
    aggregated_machine = aggregate_ioutil(aggregated_machine)
    aggregated_machine = aggregate_resptime(aggregated_machine)
    aggregated_machine = aggregate_mem(aggregated_machine)
    return aggregated_machine
```

```python
def get_batches(x_list, y_list, batch_size=50):
    '''Create a generator that returns batches of size
       n_seqs x n_steps from arr.
       
       Arguments
       ---------
       arr: Array you want to make batches from
       n_dimensions: number of features of X
       n_seqs: Batch size, the number of sequences per batch
       n_steps: Number of sequence steps per batch
    '''
    for Idx in range(0, len(x_list) - batch_size + 1, batch_size):
        batch_x = x_list[Idx:Idx + batch_size]
        batch_y = y_list[Idx:Idx + batch_size]
        yield batch_x, batch_y
```

## lstm模型

### 输入

```python
def build_inputs(batch_size, num_steps, input_size, output_size):
    ''' Define placeholders for inputs, targets, and dropout 
    
        Arguments
        ---------
        batch_size: Batch size, number of sequences per batch
        num_steps: Number of sequence steps in a batch
        
    '''
    # Declare placeholders we'll feed into the graph
    inputs = tf.placeholder(
        tf.float32, [batch_size, num_steps, input_size], name='inputs')
    targets = tf.placeholder(tf.float32, [batch_size, 1], name='targets')

    # Keep probability placeholder for drop out layers
    keep_prob = tf.placeholder(tf.float32, name='keep_prob')

    return inputs, targets, keep_prob
```

### 建立lstm

```python
def lstm_cell(lstm_size):
    return tf.contrib.rnn.BasicLSTMCell(lstm_size)

def build_lstm(lstm_size, num_layers, batch_size, keep_prob):
    ''' Build LSTM cell.
    
        Arguments
        ---------
        keep_prob: Scalar tensor (tf.placeholder) for the dropout keep probability
        lstm_size: Size of the hidden layers in the LSTM cells
        num_layers: Number of LSTM layers
        batch_size: Batch size

    '''
    ### Build the LSTM Cell
    # Use a basic LSTM cell
    lstm = tf.contrib.rnn.BasicLSTMCell(lstm_size)

    # Add dropout to the cell
    #drop = tf.contrib.rnn.DropoutWrapper(lstm, output_keep_prob=keep_prob)

    #Stack up multiple LSTM layers, for deep learning
    #cell = tf.contrib.rnn.MultiRNNCell([drop] * num_layers)
    #https://www.tensorflow.org/tutorials/recurrent
    stacked_lstm = tf.contrib.rnn.MultiRNNCell(
    [lstm_cell(lstm_size) for _ in range(num_layers)])
    initial_state = stacked_lstm.zero_state(batch_size, tf.float32)

    return stacked_lstm, initial_state
```

### 建立输出

```python
def build_output(lstm_output, in_size, output_size):
    ''' Build a softmax layer, return the softmax output and logits.
    
        Arguments
        ---------
        
        lstm_output: List of output tensors from the LSTM layer
        in_size: Size of the input tensor, for example, size of the LSTM cells
        out_size: Size of this softmax layer
    
    '''
    # 全连接层
    seq_output = tf.concat(lstm_output, axis=1)
    x = tf.reshape(seq_output, [-1, in_size])

    # Connect the RNN outputs to a softmax layer
    with tf.variable_scope('fully_conected'):
        fully_conected_w = tf.Variable(
            tf.truncated_normal((in_size, output_size), stddev=0.1))
        fully_conected_b = tf.Variable(tf.zeros(output_size))

    out = tf.matmul(x, fully_conected_w) + fully_conected_b

    return out
```

### 损失函数

```python
def build_loss(logits, targets):
    ''' Calculate the loss from the logits and the targets.
    
        Arguments
        ---------
    '''
    #y_reshaped = tf.reshape(targets, (batch_size,1))
    loss = tf.losses.mean_squared_error(predictions=logits, labels=targets)
    return loss
```

### 优化器

```python
def build_optimizer(loss, learning_rate, grad_clip):
    ''' Build optmizer for training, using gradient clipping.
    
        Arguments:
        loss: Network loss
        learning_rate: Learning rate for optimizer
    
    '''

    # Optimizer for training, using gradient clipping to control exploding gradients
    tvars = tf.trainable_variables()
    grads, _ = tf.clip_by_global_norm(tf.gradients(loss, tvars), grad_clip)
    train_op = tf.train.AdamOptimizer(learning_rate)
    optimizer = train_op.apply_gradients(zip(grads, tvars))

    return optimizer
```

### RNN

```python
class SL_RNN:
    def __init__(self,
                 batch_size,
                 num_steps,
                 lstm_size,
                 num_layers,
                 learning_rate,
                 input_size,
                 output_size,
                 grad_clip=5,
                 sampling=False,):

        # When we're using this network for sampling later, we'll be passing in
        # one character at a time, so providing an option for that
        tf.reset_default_graph()

        # Build the input placeholder tensors
        self.inputs, self.targets, self.keep_prob = build_inputs(
            batch_size, num_steps, input_size, output_size)

        # Build the LSTM cell
        cell, self.initial_state = build_lstm(lstm_size, num_layers,
                                              batch_size, self.keep_prob)

        # Run each sequence step through the RNN and collect the outputs
        outputs, state = tf.nn.dynamic_rnn(
            cell, inputs=self.inputs, initial_state=self.initial_state)
        self.final_state = state

        # Get softmax predictions and logits
        self.prediction = build_output(outputs[:, -1], lstm_size, output_size)

        # Loss and optimizer (with gradient clipping)
        self.loss = build_loss(self.prediction, self.targets)
        self.optimizer = build_optimizer(self.loss, learning_rate, grad_clip)
```

## 超参数

```python
batch_size = batch_size  # Sequences per batch
num_steps = TIME_STEP  # Number of sequence steps per batch
lstm_size = 50  # Size of hidden layers in LSTMs
num_layers = 3  # Number of LSTM layers
learning_rate = 0.0001  # Learning rate
keep_prob = 1  # Dropout keep probability
epochs = 50
# Save every N iterations
save_every_n = 200
```

## 建立模型

```python
model = SL_RNN(
    batch_size=batch_size,
    num_steps=num_steps,
    lstm_size=lstm_size,
    num_layers=num_layers,
    learning_rate=learning_rate,
    input_size=INPUT_SIZE,
    output_size=1
)

saver = tf.train.Saver()
```

## 开始训练

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    # Use the line below to load a checkpoint and resume training
    #saver.restore(sess, 'checkpoints/______.ckpt')
    counter = 0
    loss = []
    for e in range(epochs):
        # Train network
        
        new_state = sess.run(model.initial_state)
        
        for x, y in get_batches(x_list, y_list, batch_size=batch_size):
            counter += 1
            start = time.time()
            feed = {
                model.inputs: x,
                model.targets: y,
                model.keep_prob: keep_prob,
                model.initial_state: new_state
            }
            batch_loss, new_state, _ = sess.run(
                [model.loss, model.final_state, model.optimizer],
                feed_dict=feed)
            
            loss.append(batch_loss)
            end = time.time()
            print('Epoch: {}/{}... '.format(e + 1, epochs), 
                  'Training Step: {}... '.format(counter),
                  'Training loss: {:.4f}... '.format(batch_loss),
                  '{:.4f} sec/batch'.format((end - start)))
            if (counter % save_every_n == 0):
                saver.save(sess, "checkpoints/i{}_l{}.ckpt".format(
                    counter, lstm_size))

    saver.save(sess, "checkpoints/i{}_l{}.ckpt".format(counter, lstm_size))
```
